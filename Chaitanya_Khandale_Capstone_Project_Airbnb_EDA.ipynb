# **Project Summary -**

Exploring Airbnb data through EDA is beneficial for a wide array of stakeholders such as hosts, property managers, analysts, researchers, and Airbnb itself. Here are the rephrased objectives for conducting an EDA:

* Assessing Data Accuracy: EDA is instrumental in verifying the precision of Airbnb's data by identifying issues like absent values, anomalies, and data discrepancies, which is critical for robust analysis.

* Insight into Pricing Decisions: EDA provides insights that can guide hosts and property managers on pricing strategies. This includes adjusting nightly rates based on historical pricing data and understanding seasonal demand to enhance revenue.

* Enhancing Guest Experiences: EDA enables hosts to pinpoint which aspects of their listings are most impactful in delivering a gratifying guest experience. This might include evaluating factors such as amenities, cleanliness, and interaction with guests.

* Gleaning Market Understanding: EDA helps investors and property managers grasp the dynamics of the local Airbnb market. Insights into the mix of properties, their distribution, and occupancy rates can inform real estate purchase and investment decisions.

* Identifying Seasonal Patterns: EDA reveals patterns in bookings across different seasons, aiding hosts in tailoring their strategies for peak and off-peak periods, which might involve price adjustments or promotional efforts.

* Uncovering Competitive Edges: EDA can help hosts discover what sets their offerings apart from the competition. This may involve analyzing unique amenities, locations, or features that attract guests.

* Informing Marketing Efforts: EDA provides data on Airbnb guest demographics and preferences, which can be leveraged for targeted marketing initiatives aimed at specific traveler segments.

* Analyzing Reviews for Enhanced Performance: Sentiment analysis and EDA of user reviews can unveil prevalent sentiments and trends, guiding hosts in addressing areas of concern and enhancing guest satisfaction.

* Ensuring Regulatory Adherence: EDA is a tool for hosts and property managers to align with local regulations, a necessity in jurisdictions with specific Airbnb rules.

* Fueling Academic Inquiry: For researchers and academics, EDA of Airbnb data can shed light on the impacts of short-term rentals on housing markets, local economies, and tourism, and can be pivotal in studies of gentrification and housing affordability.

* Advancing Platform Development: Airbnb can utilize EDA to delve into user interactions, behaviors, and feedback, which can drive platform improvements, policy updates, and strategic marketing decisions.

* Empowering Data-Driven Choices: At its core, the primary aim of Airbnb EDA is to facilitate informed decision-making grounded in data. Whether it's for pricing, guest experience, or strategic investments, EDA lays the foundation for informed decisions within the Airbnb ecosystem.

##**Which neighborhoods have the most Airbnb listings within each neighborhood group?**

```python
# Group by "neighbourhood_group" and then count listings in each "neighbourhood" within the group
area_df = new_df.groupby("neighbourhood_group")["neighbourhood"].value_counts().reset_index(name='listing_count')
area_df = area_df.sort_values(by='listing_count', ascending=False)
pd.DataFrame(area_df)
```

```python
top_neighborhoods_df = area_df.groupby("neighbourhood_group").head(5)

# Pivot the DataFrame for easier plotting
pivot_df = top_neighborhoods_df.pivot(index="neighbourhood", columns="neighbourhood_group", values="listing_count")

# Create a horizontal bar chart using Pandas plot
ax = pivot_df.plot(kind="barh", figsize=(12, 8), color=plt.cm.plasma(np.linspace(0, 1, 5)), edgecolor='black')

# Adding data labels on each bar
for i, group in enumerate(pivot_df.columns):
    for j, value in enumerate(pivot_df[group]):
        ax.text(value, j, f'{value}', ha='left', va='center', color='darkblue', fontsize=10)

# Adding labels and title
ax.set_xlabel('Number of Listings')
ax.set_ylabel('Neighborhood')
ax.set_title('Top Neighborhoods by Number of Listings within Each Group')

# Adding legend with neighborhood group names
ax.legend(title='Neighborhood Group', loc='upper right')

# Adding grid lines
ax.grid(axis='x', linestyle='--', alpha=0.6)

# Invert y-axis for better readability
ax.invert_yaxis()

plt.tight_layout()
plt.show()
```
